```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Declaração de variáveis do problema
Vmax = 1
f = 60
R=10

# Tensão  de entrada
n = 1000000
x = np.linspace(0, 1/f, n)
Vin = Vmax*np.sin(2*np.pi*f*x)
```

```python
# Tensão e corrente na carga
Vo = np.abs(Vin)
Vmed = 1/n * sum(Vo)

ic = Vo/R
iRMS = np.sqrt(1/n*np.sum(np.square(ic)))
imed = Vmed/R

print("Tensão média na carga:", Vmed)
print("Valor eficaz da corrente na carga:", iRMS)
```

```python
plt.subplot(211)
plt.plot(x,Vin, label = 'Vin')
plt.grid(True)
plt.title('Tensão de entrada')
plt.xlabel("t [s]")
plt.ylabel("Vin [V]")
plt.legend()
plt.tight_layout()

plt.subplot(212)
plt.plot(x,Vo, label = "Tensão na carga", linewidth = 1)
plt.plot(x,ic, color = 'r', label = "Corrente na carga")
plt.axhline(Vmed, color='b', label=f'VCC = {Vmed:.4f}', linestyle='--')
plt.grid(True)
plt.title('Tensão na carga')
plt.xlabel("t [s]")
plt.ylabel("V [V]")
plt.legend()
plt.tight_layout()
```

```python
rippleI = ( np.max(np.abs(ic)) - np.min(np.abs(ic)) ) / imed
rippleV = ( np.max(np.abs(Vo)) - np.min(np.abs(Vo)) ) / Vmed

print("Ripple da corrente na carga:", rippleI)
print("Ripple da tensão na carga:", rippleV)
```
